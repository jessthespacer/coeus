```python
from scholarly import scholarly
import pandas as pd
```

```python
def getAvgCiteRate(pubs):
    rateDict = pubs.cites_per_year
    rateList = [rateDict[key] for key in rateDict]
    return sum(rateList)/len(rateList)

def getNumPubs(dfRow, debug=False):
    prof = dfRow['Name']
    
    try:
        # Query Google Scholar for the prof and get their profile
        search_query = scholarly.search_author(prof)
        author = next(search_query)

        # Get publication stats
        pubs = author.fill(sections=['counts'])

        # Calculate citation rate [cites/year]
        citeRate = getAvgCiteRate(pubs)
    except:
        # If this runs into an error, set rate to -1 to indicate
        # that something went wrong
        citeRate = -1
    if debug:
        print(prof, citeRate)
    return citeRate
```

```python
profs = pd.read_csv('profnames.txt')
```

```python
# Run queries and add average cite rate to new column
profs['avgCiteRate'] = profs.apply(getNumPubs, axis=1)
```
