Searches Google for faculty pages.

```python
from googlesearch import search
import pandas as pd

# Import university dataframe
data = pd.read_csv('uni_data.csv')

# Grab first 10 only
data = data[0:10]
```

```python
data
```

```python
# General search terms
faculties = ['engineering', 'physics', 'astronomy', 'arts', 'biology', 'chemistry']
targetTerms = ['people', 'faculty', 'professors', 'group']
targetTerms = ' OR '.join(targetTerms)

def getStaffPages(dfRow, faculty):
    query = faculty + ' ' + targetTerms + ' site:' + dfRow['url']
    results = [x for x in search(query, lang = 'en', start = 0, stop = 20, pause = 2.0)]
    return results

getStaffPagesForFaculty = lambda dfRow: getStaffPages(dfRow, faculties[1])

for i, row in data.iterrows():
    try:
        row['search results'] = getStaffPagesForFaculty(row)
    except:
        print('Probably error 429')
```
